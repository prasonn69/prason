```python
# Concatenate the string Python, 4, Data, Science to a single string, Python 4 Data Science
# Declare a variable named course and assign it to an initial value Python 4 Data Science
course = "Python 4 Data Science"

# Print the length of the course string using len() method and print()
print(len(course))  

# Change all the characters of variable company to uppercase and lowercase letters using upper() and lower() method
print(course.upper())    
print(course.lower())    

# Use capitalize(), title(), swapcase() methods to format the value of the string Python 4 Data Science
print(course.capitalize())  
print(course.title())      
print(course.swapcase())   

# Cut(slice) out the first word of Python 4 Data Science
print(course[:6])  

# Check if Python 4 Data Science string contains a word Python using the method: index(), find() or other methods
print(course.index("Python"))  
print(course.find("Python"))   

# Change Python 4 Data Science to Python 4 Everybody using the replace() method or other methods
print(course.replace("Data Science", "Everybody"))  

# Split the string Python 4 Data Science using space as the separator (split())
print(course.split(" "))  

# Google, Facebook, Microsoft, Apple, IBM, Oracle, Amazon split the string at the comma
companies = "Google, Facebook, Microsoft, Apple, IBM, Oracle, Amazon"
print(companies.split(", "))  

# What is the character at index 9 in the string Python 4 Data Science
print(course[9])          

# What is the second last index of the string Python 4 Data Science
print(course[-2])        

# Create an acronym or an abbreviation for the name Python 4 Data Science
acronym = "".join(word[0] for word in course.split())  # P4DS
print(acronym)

# Use index() to determine the position of the first occurrence of D in Python 4 Data Science
print(course.index("D"))         

# Use rfind to determine the position of the last occurrence of e in Python 4 Data Science
print(course.rfind("e"))          

```
